# Main Code

```python
import cvxpy as cp
import pandas as pd
import numpy as np

df = pd.read_csv('FINAL_SHEET.csv')
is_word = pd.read_csv('VALID_WORDS.csv')

constraints = []

f = cp.Variable(5, integer = True)

scores = []

obj_func = f[0] + f[1] + f[2] + f[3] + f[4]

constraints.append(f[0] >=  0)
constraints.append(f[1] >=  0)
constraints.append(f[2] >=  0)
constraints.append(f[3] >=  0)
constraints.append(f[4] >=  0)

print ("Welcome to Wordle Guesser!")
print ("Lets start by entering your initial word:")
print("")

yellows_arr = []

yellows_dict = {
    0:"-",
    1:"-",
    2:"-",
    3:"-",
    4:"-"
}

greens_dict = {
    0:"-",
    1:"-",
    2:"-",
    3:"-",
    4:"-"
}

greens = ["-", "-", "-", "-", "-"]
yellows = ["-", "-", "-", "-", "-"]
greys = ["-", "-", "-", "-", "-"]

def find_keys(dictionary, target_value):
    return [key for key, value in dictionary.items() if value == target_value]



a = 0;
while (a != 5):

    ## Create a save of the dataframe
    df_save = df.copy()

    print("Please enter all lowercase letters: ")
    print("")

    for i in range(5):
        green = input("If the letter in position " + str(i+1) + " is green, please enter it below (enter '-' if it isn't)")

        if(green != '-'):
            greens[i] = green
            greens_dict[i] = green
            df = df[df['word'].str[i] == green]

    print("")

    for i in range(5):
        yellow = input("If the letter in position " + str(i+1) + " is yellow, please enter it below (enter '-' if it isn't)")

        if(yellow != '-'):
            yellows[i] = yellow
            yellows_dict[i] = yellow
            yellows_arr.append(yellow)
            df = df[df['word'].apply(lambda x: len(x) > 1 and x[i] != yellow)]

    print("")

    for i in range(5):

        grey = input("If the letter in position " + str(i+1) + " is grey, please enter it below (enter '-' if it isn't)")

        if(grey != '-'):
            greys[i] = grey

            ## FIX THE GREEN VALUES
            if grey in greens_dict.values():
                ## Remove all the greys from every character except the character the green is at
                ## Figure out which character the green is at
                temp = find_keys(greens_dict,grey)
                temp2 = [0,1,2,3,4]
                filtered_array = [element for element in temp2 if element not in temp]

                for d in filtered_array:
                    df = df[df['word'].str[d] != grey]

            ## FIX THE YELLOW VALUES
            elif grey in yellows_dict.values():
                ## Figure out what positions yellow is not allowed to be
                df = df[df['word'].apply(lambda x: len(x) > 1 and x[i] != grey)]

            else:
                df = df[~df['word'].apply(lambda x: grey in x)]


    print("")

    for i in range(5):
        if(greens[i] != "-" or yellows[i] != "-" or greys[i] != "-"):
            a = a + 1

    if (a != 5):
        print("ERROR- Each character must be green, yellow, or grey. Please reinput all characters to meet requirements")
        print("")
        print("")

        ## Redefine the dataset and restart the loop
        df = df_save
        a = 0

## PROMPT USER TO REMOVE PREVIOUS GUESSES
print("")
user = input("How many words have you previously guessed? ")
print("")
print("")
guessed = int(user)


## NEED TO TWEAK THIS 
for i in range(guessed):
    
    statement = "Lets get some info for word number "
    temp_int = i + 1
    print(statement, temp_int)
    
    print("")
    print("")

    yellows_dict = {
        0:"-",
        1:"-",
        2:"-",
        3:"-",
        4:"-"
    }

    greens_dict = {
        0:"-",
        1:"-",
        2:"-",
        3:"-",
        4:"-"
    }

    greens = ["-", "-", "-", "-", "-"]
    yellows = ["-", "-", "-", "-", "-"]
    greys = ["-", "-", "-", "-", "-"]

    def find_keys(dictionary, target_value):
        return [key for key, value in dictionary.items() if value == target_value]



    a = 0;
    while (a != 5):

        ## Create a save of the dataframe
        df_save = df.copy()
        
        print("Please enter all lowercase letters: ")
        print("")

        for i in range(5):
            green = input("If the letter in position " + str(i+1) + " is green, please enter it below (enter '-' if it isn't)")

            if(green != '-'):
                greens[i] = green
                greens_dict[i] = green
                df = df[df['word'].str[i] == green]

        print("")

        for i in range(5):
            yellow = input("If the letter in position " + str(i+1) + " is yellow, please enter it below (enter '-' if it isn't)")

            if(yellow != '-'):
                yellows[i] = yellow
                yellows_dict[i] = yellow
                df = df[df['word'].apply(lambda x: len(x) > 1 and x[i] != yellow)]

        print("")

        for i in range(5):

            grey = input("If the letter in position " + str(i+1) + " is grey, please enter it below (enter '-' if it isn't)")

            if(grey != '-'):
                greys[i] = grey

                ## FIX THE GREEN VALUES
                if grey in greens_dict.values():
                    ## Remove all the greys from every character except the character the green is at
                    ## Figure out which character the green is at
                    temp = find_keys(greens_dict,grey)
                    temp2 = [0,1,2,3,4]
                    filtered_array = [element for element in temp2 if element not in temp]

                    for d in filtered_array:
                        df = df[df['word'].str[d] != grey]

                ## FIX THE YELLOW VALUES
                elif grey in yellows_dict.values():
                    ## Figure out what positions yellow is not allowed to be
                    df = df[df['word'].apply(lambda x: len(x) > 1 and x[i] != grey)]

                else:
                    df = df[~df['word'].apply(lambda x: grey in x)]


        print("")

        for i in range(5):
            if(greens[i] != "-" or yellows[i] != "-" or greys[i] != "-"):
                a = a + 1

        if (a != 5):
            print("ERROR- Each character must be green, yellow, or grey. Please reinput all characters to meet requirements")
            print("")
            print("")

            ## Redefine the dataset and restart the loop
            df = df_save
            a = 0
    
    
        
j = cp.Variable(len(df), boolean = True)
constraints.append(sum(j) == 1)   

for i in range(len(df)):

    multiplier = 1;

    if df.iat[i,0][0] in yellows_arr:
        multiplier += 10;
    if df.iat[i,0][1] in yellows_arr:
        multiplier += 10;
    if df.iat[i,0][2] in yellows_arr:
        multiplier += 10;
    if df.iat[i,0][3] in yellows_arr:
        multiplier += 10;
    if df.iat[i,0][4] in yellows_arr:
        multiplier += 10;


    constraints.append(f[0] <= df.iat[i,1] * multiplier + ((1 - j[i]) * 100000))
    constraints.append(f[0] >= df.iat[i,1] * multiplier - ((1 - j[i]) * 100000))


    constraints.append(f[1] <= df.iat[i,2] * multiplier + ((1 - j[i]) * 100000))
    constraints.append(f[1] >= df.iat[i,2] * multiplier - ((1 - j[i]) * 100000))


    constraints.append(f[2] <= df.iat[i,3] * multiplier + ((1 - j[i]) * 100000))
    constraints.append(f[2] >= df.iat[i,3] * multiplier - ((1 - j[i]) * 100000))


    constraints.append(f[3] <= df.iat[i,4] * multiplier + ((1 - j[i]) * 100000))
    constraints.append(f[3] >= df.iat[i,4] * multiplier - ((1 - j[i]) * 100000))


    constraints.append(f[4] <= df.iat[i,5] * multiplier + ((1 - j[i]) * 100000))
    constraints.append(f[4] >= df.iat[i,5] * multiplier - ((1 - j[i]) * 100000))

    scores.append(multiplier * (df.iat[i,1] + df.iat[i,2] + df.iat[i,3] + df.iat[i,4] + df.iat[i,5]))

    
print("")
print("PROBLEM SOLVING... PLEASE WAIT...")
    
print("")    

problem = cp.Problem(cp.Maximize(obj_func), constraints)
problem.solve(solver=cp.GUROBI, verbose = False)

print("")
print("")


# Print out the guessed word
print("Ideal Guess:")        
for i in range(len(scores)):
    if(scores[i] == obj_func.value):
        print(df.iat[i, 0])

print("")
print("")


# Print out all of the possible next guesses given the constraints 
is_word = pd.read_csv('VALID_WORDS.csv')
print("Number of Potential Guesses:")
print(len(df))
print("")
print("")
print("Potential Guesses:")

potential_guesses = pd.DataFrame(columns = ['Word', 'Is_Word', 'Frequency_Score'])

for i in range(len(df)):
    word = df.iat[i,0]
    sum = df.iat[i,1] + df.iat[i,2] + df.iat[i,3] + df.iat[i,4] + df.iat[i,5]
    word_intro = "Word: "
    sum_intro = "   Frequency Score: "
    is_word_intro = "   Is a Solution:"
    is_in_column = is_word['word'].isin([word])
    is_in_column = is_in_column.any()
    print(word_intro, word, sum_intro, sum, is_word_intro, is_in_column)
    
print(potential_guesses.to_string(index = False))

# Print the objective values
print("")
print("Objective Function: ")
print(obj_func.value)
print("")

print("Frequency Values: ")
print(f[0].value, f[1].value, f[2].value, f[3].value, f[4].value)
print("")
```
